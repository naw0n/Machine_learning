**특성 공학** : 기존의 특성을 사용해 새로운 특성을 뽑아내는 작업

ex) 농어 길이 * 농어 높이

데이터프레임 : pandas의 핵심 데이터 구조

csv : ,로 나누어져 있는 텍스트 파일

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures()
# 2,3 2개의 특성으로 이뤄진 샘플 적용
poly.fit([[2,3]])
print(poly.transform([[2,3]])) # 샘플 [2,3] -> 6개의 특성을 가진 [1,2,3,4,6,9] 로 변환
```

**변환기** : 특성을 만들거나 전처리 하기 위한 다양한 클래스

sklearn model 클래스 : 일관된 fit(), score(), predict()

변환기 클래스 : 일관된 fit(), transform()
- fit : 새롭게 만들 특성 조합 찾음
- transform : 실제로 데이터 변환

transform 사용 전 fit 사용 = 훈련을 해야 변환 가능

fit_transform 메소드 : 두 메소드를 하나도 붙임

PolynomialFeatures 클래스 : 기본적으로 각 특성을 제곱한 항 추가, 특성끼리 서로 곱한 값 추가
특성 ( 길이, 높이, 두께, 1) 

무게 = a * 길이 + b * 높이 + c * 두께 + 1 sklearn의 선형모델은 자동으로 절편을 추가하기 때문에 굳이 1을 넣어 특성을 만들 필요는 없다

```python
poly = PolynomialFeatures(include_bias=False) # 1 제거 -> 특성의 제곱과, 특성끼리 곱한 항만 추가
poly.fit([[2,3]])
print(poly.transform([[2,3]]))
```

```python
poly.fit(train_input)
train_poly = poly.transform(train_input)
print(train_poly.shape) # 배열의 크기 확인
```

```python
poly.get_feature_names_out()
```

get_feature_names_out : 특성이 어떤 입력의 조합으로 만들어졌는지

```python
test_poly = poly.transform(test_input)
```

변환된 특성을 사용하여 다중 회귀 모델 훈련

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
```

 과소적합 문제 해결

```python
print(lr.score(test_poly, test_target))
```

테스트 세트에 대한 점수는 변동 없음

```python
poly = PolynomialFeatures(degree=5, include_bias=False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
print(train_poly.shape)
```

특성을 더 추가하여 3제곱, 4제곱 항을 추가면 특성의 개수가 55개로 증가한다. (train_poly의 열의 개수 = 특성 개수)

```python
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))
```

### 알 수 있는 점
특성 개수 증가 -> 선형 모델은 강력해짐 = 훈련 세트에 대해 거의 완벽하게 학습 가능

그러나, 훈련 세트에 *과대 적합*
